```
import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('/content/drive/My Drive/Keras/LearnKeras-master/1.LinearRegression/IowaHousingPrices.csv')
```

```
df.head()
```

```
sqaureFeet = df[['SquareFeet']].values
```

```
salePrice = df[['SalePrice']].values
```

```
model = keras.Sequential()
model.add(keras.layers.Dense(1, input_shape=(1,)))
```

```
model.compile(keras.optimizers.adam(lr=1),'mean_squared_error')
model.fit(sqaureFeet,salePrice, epochs=30)
```

```
df.plot(kind='scatter', x='SquareFeet', y='SalePrice')
y_pred = model.predict(sqaureFeet)

plt.plot(sqaureFeet,y_pred,color = 'red')
```
