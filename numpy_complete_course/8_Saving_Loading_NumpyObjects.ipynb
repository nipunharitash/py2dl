# Saving and Loading Numpy Objects

```python
import numpy as np
from numpy import random
```

```python
a1 = np.array([[1,2,3],[4,5,6]])
a1
```

```python
np.save('savea1', a1)
```

```python
a2 = np.load('savea1.npy')
```

```python
a2
```

```python
a3 = np.random.randint(1, 20, size=(100,3))
```

```python
a3
```

```python
np.savetxt('savecsva3.csv', a3)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('savecsva3.csv')
df
```

```python
a4 = np.loadtxt('savecsva3.csv')
```
