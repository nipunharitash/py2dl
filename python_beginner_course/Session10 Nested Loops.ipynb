```python
num1 = 25
num2 = 50
for i in range(num1,num2+1):
    if num1>1:
        for j in range(2,num1):
            if(i%j==0):
                break
        else:
            print(i)
```

```python
def prime(num1,num2):
    for i in range(num1,num2+1):
        if num1>1:
            for j in range(2,num1):
                if(i%j==0):
                    break
                else:
                    print(i)
```

```python
prime(25,50)
```
